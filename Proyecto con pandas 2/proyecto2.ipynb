```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python

lego_set = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
lego_parent_theme = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
lego_set.shape
```

#### Eliminar los valores nulos de los datos en la coluna "set_num"

```python
lego_set = lego_set[lego_set["set_num"].notna()]
lego_set
```

```python
# What porcentages of all instance sets ever released were Star Wars Themed
#merge datas 
all_data = lego_set.merge(lego_parent_theme,left_on="parent_theme",right_on="name")
all_data.drop(columns="name_y",inplace=True)
all_data.head()
licensed =all_data[all_data["is_licensed"]]
all_data
star_wars = licensed[licensed["parent_theme"] == "Star Wars"]
the_force = star_wars.shape[0]/licensed.shape[0]*100
the_force
```

```python
#in which year was star wars not the most popular licensed theme

licensed_group=licensed.sort_values("year").groupby(["year","parent_theme"]).sum().reset_index()
licensed_group.sort_values("is_licensed",ascending=False).drop_duplicates(["year"])

```
